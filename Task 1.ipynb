```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords

# Sample Data: Create a DataFrame to simulate your data
data = {
    'name': ["Jonathan Mallett", "Aditya Rawale", "Julien Ranke", "Tommy smith", "Amanda Newell"],
    'text': [
        "Really needs an option to copy single items from meals across days instead of being forced to copy entire meals.",
        "not good",
        "Excellent app to track what you're eating!",
        "I have been using this since 2014, I recommend the app to anyone with all kinds of different goals.",
        "love it"
    ],
    'stars': [2, 2, 5, 5, 5],
    'category': ["Health & Fitness", "Health & Fitness", "Health & Fitness", "Health & Fitness", "Health & Fitness"]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Filter for 5-star reviews in the "Health & Fitness" category
filtered_reviews = df[(df['stars'] == 5) & (df['category'] == "Health & Fitness")]

# Combine all reviews into a single string
text = " ".join(filtered_reviews['text'])

# Define stopwords
stop_words = set(stopwords.words('english'))  # Use NLTK's English stopwords
# Add additional app-specific stopwords
app_specific_stopwords = ['app', 'health', 'fitness', 'track', 'tracking', 'goal', 'goals', 'using', 'recommend', 'love']
stop_words.update(app_specific_stopwords)

# Generate word cloud
wordcloud = WordCloud(stopwords=stop_words, background_color='white', width=800, height=400).generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Turn off axis
plt.show()
```
